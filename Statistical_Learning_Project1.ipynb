```python
def Goal_diff_count(Teams):
    return int(Teams[1])-int(Teams[2])

GoalDiff = laliga[['Team','GoalsFor','GoalsAgainst']].apply(lambda x: Goal_diff_count(x),axis=1)
```

```python
print("The teams and their goal differences are as follows:")
for x in range(GoalDiff.count()):
    print(laliga.loc[x,'Team']+" goaldiff is "+str(GoalDiff[x]))
```

```python
MinGD = GoalDiff[0:60].min()
MaxGD = GoalDiff[0:60].max()
```

```python
MinIN=GoalDiff[GoalDiff==MinGD].index
MaxIN=GoalDiff[GoalDiff==MaxGD].index
```

```python
print(laliga.loc[MinIN,'Team'])
print(laliga.loc[MaxIN,'Team'])
```

```python
def Win_Percent(P):
    if int(P[1])!=0:
        return int(P[0])/int(P[1])
    else:
        return 0
laliga['Winning Percent'] = laliga[['GamesWon','GamesPlayed']].apply(lambda x: Win_Percent(x),axis=1)
```

```python
laliga
```

```python
indexpr = laliga['Winning Percent'].astype(str).astype(float).nlargest(5).index
indexpr
```

```python
print("The top 5 teams with higest points are as follows:")
Top5 = laliga.iloc[indexpr]
Top5['Team']
```
