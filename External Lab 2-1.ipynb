### Find out the no of male & female employees in the company

*Hint: Use can use a condition to find out the no of 1s and 0s in gender and sum it*

```python
Male = dataset[dataset['Gender']==1]['Gender'].count()
Female = dataset[dataset['Gender']==2]['Gender'].count()
```

```python
print("No of male employees ="+str(Male))
print("No of female employees ="+str(Female))
```

### Find out the IQR for the column - Age 

*Hint: IQR - Interquartile range = 75th percentile - 25th percentile* 

```python
IQR = dataset['Age'].quantile(0.75) - dataset['Age'].quantile(0.25)
print(IQR)
```
